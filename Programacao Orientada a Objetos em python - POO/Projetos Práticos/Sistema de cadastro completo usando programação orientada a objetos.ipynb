```python
# Criando uma classe CRUD (Create, Read, Update, Delete)
class CRUD:
    
    # Definindo um método construtor
    def __init__(self):
        import os
        self.os = os # Definindo um atributo para instanciar a classe os
        os.system('cls') # Limpa a tela do terminal na criação da classe
        print('CADASTRO') # Mensagem sempre exibida ao iniciar a execução da classe CRUD
        
    # Função para ler os dados cadastrados        
    def ler(self):
        
        # Verifica se o arquivo base_dados.txt ainda não existe dentro do diretório atual
        if not self.os.path.exists("base_dados.txt"):
            
            escrita = open('base_dados.txt', 'w') # Configura a criação do arquivo base_dados.txt para escrita
            escrita.write('') # Executa a criação do arquivo sem dados ainda
            escrita.close() # Fecha o arquivo para escrita (libera da memória)
        
        leitura = open("base_dados.txt", 'r') # configura o arquivo base_dados.txt para apenas leitura
        retorno = ''
        
        # Captura cada linha de texto (cada cadastro) do arquivo base_dados.txt
        for linha in leitura:
            retorno += linha.strip() + '\n' # Elimina os espaços no início e fim de cada linha e adiciona ao objeto retorno
        
        leitura.close() # Fecha o arquivo de leitura (libera da memória)
        
        if len(retorno) > 0: return retorno # Se o objeto retorno contiver cadastros a serem lidos, retorna o conteúdo do objeto
        
        else: return 'Nenhum registro cadastrado' # Caso não haja cadastros a serem lidos, essa mensagem é exibida na tela
        
    # Função para gerar o id (número de registro)
    def codigo_id(self):
        self.ler() # Faz a leitura do arquivo base_dados.txt. Caso ainda não exista, ele será criado.
        leitura = open('base_dados.txt', 'r') # Configura o arquivo para leitura
        ultima_linha = '' # Essa variável só armazenará a última linha do arquivo de leitura
        
        # Captura cada linha de texto (cada cadastro) do arquivo base_dados.txt
        for linha in leitura:            
            ultima_linha = linha.strip() # Elimina os espaços no início e fim da linha de texto do arquivo de leitura
        
        # Caso última_linha não apresente dados, última_linha receberá o id abaixo.
        if len(ultima_linha) <= 0: ultima_linha = 'ID: 0000000'
        
        # Captura apenas os números na string 'ID: 0000000' e soma 1 para incrementar o id
        id = int(ultima_linha[ultima_linha.find('ID: ') + 4: 11]) + 1
        
        # Sempre haverá 7 caracteres para cada id. A função abaixo adiciona 0's a esquerda de cada valor para o id.
        id = str(id).rjust(7, '0')
        
        leitura.close() # Fecha o arquivo de leitura (libera da memória)
        
        return id # Retorna o id
    
    # Método para a inserção de dados
    def cadastrar(self, nome = '', idade = ''):
        
        id = self.codigo_id() # O objeto id recebe o valor do id criado e retornado pela função acima
        nome = str(nome).ljust(25, ' ') # Serve para garantir que o nome estará no formato string e justifica a esquerda
        idade = str(idade).rjust(3, '0') # Adiciona 0's a esquerda do valor para a idade até o valor apresentar 3 caracteres
        
        texto = self.ler() # Abre o arquivo base_dados.txt configurado para leitura
        
        # Caso o objeto texto que instancia retorno não apresente nenhum cadastro, texto receberá uma string vazia
        if texto == 'Nenhum registro cadastrado': texto = ''
        
        escrita = open('base_dados.txt', 'w') # Abre o arquivo base_dados.txt p/ escrita
        linha = f'ID: {id} - NOME: {nome} - IDADE: {idade} anos' # Linha para inserção no arquivo base_dados.txt
        
        escrita.write(texto + linha) # Adiciona a linha criada anteriormente ao arquivo
        escrita.close() # Libera o objeto escrita da memória
        
    # Método para alteração de dados
    def alterar(self, id = -1, nome = '', idade = ''):
        
        # Verifica se o id tem menos de 7 caracteres e, caso afirmativo, adiciona 0's a esquerda do valor do id.
        if len(id) < 7: id = str(id).rjust(7, '0')
            
        print('ESSE É O ID: ', id)
            
        nome = str(nome).ljust(25, ' ') # Adiciona 25 espaços a direita do valor para o nome.
        idade = str(idade).rjust(3, '0') # Adiciona 3 0's a esquerda do valor para a idade.
        
        texto = self.ler() # Faz a leitura do arquivo base_dados.txt. Caso ainda não exista, ele será criado.
        linhas = texto.split('\n') # Quebra as linhas de texto na quebra de linha entre as linhas
        escrita = open('base_dados.txt', 'w') # Abre o arquivo base_dados.txt p/ escrita
        texto = '' # Irá sobrescrever a variável texto criada anteriormente
        
        for linha in linhas:
            
            if linha.find(f'ID: {id}') >= 0: # Verifica se o tamanho do valor para o ID é maior ou igual a zero
                if len(nome.strip()) <= 0: # Caso o nome não seja passado pelo usuário por meio do input...
                    nome = linha[linha.find('NOME: ') + 6:linha.find(' - IDADE')] # Captura o nome já presente no arquivo
                    
                if '000' in idade: # Caso a idade não seja passada pelo usuário por meio do input...
                    idade = linha[linha.find('IDADE: ')+7:linha.find(' anos')] # Captura a idade já presente no arquivo
                    
                texto = f'ID: {id} - NOME: {nome} - IDADE: {idade} anos\n' # Prepara a linha de texto (cadastro)
                
            else:
                
                texto += linha + '\n' # Caso id < 0, a linha já presente no arquivo permanece inalterada
                
        escrita.write(texto.strip()) # Escreve o texto (os dados cadastrados) no arquivo base_dados.txt
        escrita.close() # Fecha o arquivo para liberá-lo da memória
    
    # Método para exclusão de registros
    def deletar(self, id = -1):
        
        if len(id) < 7: # Verifica se o id tem menos de 7 caracteres            
            id = str(id).rjust(7, '0') # Caso afirmativo, adiciona zeros a direita do id até completar 7 caracteres no total
            
        texto = self.ler() # Faz a leitura do arquivo base_dados.txt. Caso ainda não exista, ele será criado.
        linhas = texto.split('\n') # Quebra as linhas de texto na quebra de linha entre as linhas
        escrita = open('base_dados.txt', 'w') # Abre o arquivo base_dados.txt p/ escrita
        texto = ''
        for linha in linhas:
            
            if linha.find(f'ID: {id}') >= 0: # Caso o id correto seja passado pelo usuário por meio do input... 
                texto += '' # Não acrescenta nada no objeto texto e isso será usado para apagar o registro no arquivo
                
            else:
                texto += linha + '\n' # Caso contrário, o registro original permanece no arquivo base_dados.txt
                
        escrita.write(texto.strip()) # Escreve o texto dentro do arquivo
        escrita.close() # Libera o objeto da memória
            
    # Função para consultar registros 
    def consultar(self, id=-1):
        
        if len(id) < 7: # Verifica se o id tem menos de 7 caracteres            
            id = str(id.rjust(7, '0')) # Caso afirmativo, adiciona zeros a direita do id até completar 7 caracteres no total
        
        # Verifica se o arquivo base_dados.txt ainda não existe dentro do diretório atual
        if not self.os.path.exists('base_dados.txt'):
            escrita = open('base_dados.txt', 'w') # Configura a criação do arquivo base_dados.txt para escrita
            escrita.write('') # Executa a criação do arquivo sem dados ainda
        
        leitura = open('base_dados.txt', 'r') # Configura o arquivo para leitura
        retorno = '' # Armazenará registros presentes no arquivo
        
        for linha in leitura:
            if linha.find(f'ID: {id}') >= 0: # Caso o id correto seja passado pelo usuário por meio do input... 
                retorno += linha.strip() + '\n' # O objeto retorno armazena os registros presentes no arquivo
        
        leitura.close() # Fecha o arquivo para liberá-lo da memória
        return retorno # Essa função retorna o objeto retorno com os registros cadastrados no arquivo base_dados.txt
    
    # Método principal que executa o CRUD
    def executar(self):
        opcao = 1
        while opcao > 0: # Enquanto opcao for maior que 0...
            
            # Exibe algumas opções para o usuário escolher alguma            
            print('\n(1) - LER\n(2) - CADASTRA\n(3) - ALTERA\n(4) - DELETA\n(5) - CONSULTA\n(0) - SAIR')
            opcao = int(input('Digite a sua opção: '))
            
            # Dependendo da opção escolhida, alguns inputs são requeridos e uma função (ou método) será executada.
            if opcao == 1:
                self.os.system('cls') # Limpa a tela do terminal
                print('DADOS:\n' + self.ler()) # Invoca a função ler e printa o retorno na tela
            
            elif opcao == 2:
                nome = input('NOME: ')
                idade = input('IDADE: ')
                self.cadastrar(nome, idade) # Invoca o método cadastrar com os inputs passados
                self.os.system('cls') # Limpa a tela do terminal
                print('Registro CADASTRADO com SUCESSO') # Essa mensagem é exibida no final da etapa
            
            elif opcao == 3:
                id = input('ID: ')
                nome = input('NOME: ')
                idade = input('IDADE: ')
                self.alterar(id, nome, idade) # Invoca o método alterar com os inputs passados
                self.os.system('cls') # Limpa a tela do terminal
                print('Registro EDITADO com SUCESSO') # Essa mensagem é exibida no final da etapa
                
            elif opcao == 4:
                id = input('ID: ')
                self.deletar(id) # Invoca o método deletar com os inputs passados
                self.os.system('cls') # Limpa a tela do terminal
                print('Registro EXCLUÍDO com SUCESSO') # Essa mensagem é exibida no final da etapa
            
            elif opcao == 5:
                id = input('ID: ')
                self.os.system('cls') # Limpa a tela do terminal
                linhas = self.consultar(id)  # Invoca a função consultar com os inputs passados
                print('Dados:\n' + linhas) # os registros são exibidos no final da etapa
                
            else:
                print('ENCERRADO.') # Para a opção 0, essa mensagem é exibida e a execução da classe CRUD é encerrada.
```

```python
# Criando um objeto chamado crud para instanciar a classe CRUD
crud = CRUD()
crud.executar()
```
